# Merging the 3 datasets to get 1 dataset
 1. Inner joint of Reactions with Contents dataset on the content id column
 2. Inner joint of the merged data with ReactionType dataset on the Type column

```python
merged1_df = pd.merge(reactions_df, content_df, how = 'inner', on = 'Content ID')
print(merged1_df.shape)
merged1_df.head()
```

```python
merged2_df = pd.merge(merged1_df, reactionType_df, how = 'inner', on = 'Type')
print(merged2_df.shape)
merged2_df.head()
```

# Finding the top 5 popular categories

```python
category_scores = pd.DataFrame(merged2_df.groupby("Category").sum().sort_values('Score', ascending = False)["Score"])
#category_scores.head(5).to_csv(loc + "\Final_dataset.csv")
category_scores[:5]
```

```python
with pd.ExcelWriter(loc + "\Final_dataset.xlsx") as writer:
    merged2_df.to_excel(writer, sheet_name = "Cleaned_dataset")
    category_scores.to_excel(writer, sheet_name = "Popular_Categories")
    category_scores[:5].to_excel(writer, sheet_name = "Top5_Categories")
```
